```python
import numpy as np
import cv2
from skimage import filters
import matplotlib.pyplot as plt
```

```python

def create_edge_map(image1, image2):

    gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
    gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)


    diff = cv2.absdiff(gray1, gray2)

    edges = cv2.GaussianBlur(diff, (5,5), -1)

    edges = cv2.Canny(edges, 200, 200, L2gradient = True)

    return edges
```

```python
image1 = cv2.imread('../Images/Frame 6.png')
image2 = cv2.imread('../Images/Frame 5.png')
```

```python
edge_map = create_edge_map(image1, image2)
```

```python
plt.figure(figsize=(190, 70))
```

```python
plt.subplot(231)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.title('Image 1')

plt.subplot(233)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.title('Image 2')

plt.subplot(235)
plt.imshow(edge_map, cmap='gray')
plt.title('Edge Map')

plt.show()
```
