```python
import numpy as np
import cv2
from PIL import Image
```

```python
img= cv2.imread("./../output_image.png")
img_arr = np.array(img)
img_arr.shape
```

```python
leng = img_arr.shape[0]/10
bred = img_arr.shape[1]/10
leng=int(leng)
bred= int(bred)
leng,bred
```

```python
img_reszied= cv2.resize(img, (bred, leng))
img_reszied.shape
```

```python
img_array = np.array(img_reszied)
img_array.shape
```

```python
img_array
```

```python
img_gray = cv2.cvtColor(img_reszied, cv2.COLOR_BGR2GRAY)
img_gray.shape
```

```python
img_gray
```

```python
img_normalize= img_gray / 255.0
img_normalize
```

```python
img_normalize_uint8 = (img_normalize * 255).astype(np.uint8)
img_normalize_uint8
```

```python
img= Image.fromarray(img_normalize_uint8)
img.save("proprocessed.jpg")
```
